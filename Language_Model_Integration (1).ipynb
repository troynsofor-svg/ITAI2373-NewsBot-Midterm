```python
articles = [
    {'title': 'Article 1', 'text': 'This is the text of article 1.'},
    {'title': 'Article 2', 'text': 'This is the text of article 2.'},
    {'title': 'Article 3', 'text': 'This is the text of article 3.'}
]
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')

article_embeddings = model.encode([a['text'] for a in articles], show_progress_bar=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def find_similar_articles(query_text, top_k=5):
  query_embedding = model.encode([query_text])
  similarities = cosine_similarity(query_embedding, article_embeddings)[0]

  top_indices = similarities.argsort()[-top_k:][::-1]
  return [articles[i] for i in top_indices]
```

```python
from nltk.corpus import wordnet as wn

def expand_with_wordnet(query):
  syns = wn.synsets(query)
  lemmas = set()
  for s in syns:
    for lemma in s.lemmas():
      lemmas.add(lemma.name())
  return list (lemmas)
```
