```python
!pip install fasttext
import fasttext

# Download the pre-trained language identification model
!wget https://dl.fbaipublicfiles.com/fasttext/supervised-models/lid.176.bin

model = fasttext.load_model("lid.176.bin")

def detect_language(text):
    prediction = model.predict(text)
    lang = prediction[0][0].replace("__label__", "")
    confidence = prediction[1][0]
    return lang, confidence
```

```python
from transformers import MarianMTModel, MarianTokenizer

def translate(text, src='es', tgt='en'):
  model_name = f'Helsinki-NLP/opus-mt-{src}-{tgt}'
  # Corrected: tokenizer should be MarianTokenizer, model should be MarianMTModel
  tokenizer = MarianTokenizer.from_pretrained(model_name)
  model = MarianMTModel.from_pretrained(model_name)

  encoded = tokenizer(text, return_tensors="pt", padding=True)
  translated = model.generate(**encoded)
  out = tokenizer.decode(translated[0], skip_special_tokens=True)
  return out
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('distiluse-base-multilingual-cased-v2')

def embed(text):
  return model.encode(text)
```

```python
text1 = translate(article_text)
text2 = translate(article_text, src='en', tgt='es')
```

```python
similarity = util.cos_sim(embed(text1), embed(text2))
```

```python
regional_markers = {
    "us": ["Congress","White House", "FBI"],
    "mx": ["AMLO," "Pemex", "INE"],
    "cn": ["Xi Jinping," "Beinjing", "CCTV"]
}

def detect_cultural_context(text, lang):
  context = []
  for region, markers in regional_markers.items():
    if any(m.lower() in text.lower() for m in markers):
      context.append(region)
    return context
```
